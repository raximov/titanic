#### Yashab qolgan yo‘lovchilarning sonini hisoblash va diagrammada ko‘rsatish


```python
titanic_df['Survived'].value_counts().plot.bar(color=['red', 'green'])
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.show()
```

```python
# Yashab qolgan yo‘lovchilarning o‘rtacha foizi
titanic_df['Survived'].mean()
```

#### Gender bo‘yicha o‘rtacha omon qolish foizini hisoblash


```python
titanic_df.groupby(['Sex']).mean(numeric_only=True)
```

#### Gender va sinf bo‘yicha o‘rtacha omon qolish foizini hisoblash


```python
titanic_df.groupby(['Sex', 'Pclass']).mean(numeric_only=True)
```

```python
# Yosh 18 dan kichik bo'lgan yo'lovchilarni tanlab, ularning gender va sinf bo‘yicha omon qolish foizini hisoblash
titanic_df[titanic_df['Age'] < 18].groupby(['Sex', 'Pclass']).mean(numeric_only=True)
```
